### 1) total profit of all months per month

```python

#Bar Chart : I chose to use bar chart, because we want to perform a comparison of values across different subgroups of our data  
#, i.e. need to have the total profit of all months per month, where month is ordinal data.
#From a bar chart, we can see which groups are highest (the month of highest profit for exemple).
#We can also use histogram instead of bar chart 


import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/home/depinfo/Documents/DATA SCIENCE/Visualization TP-20200403/company_sales_data.csv", sep=',') 
print(df)

var = df.groupby('month_number').total_profit.sum() #grouped sum of sales at Gender level
fig = plt.figure()

ax1 = fig.add_subplot(1,1,1)
ax1.set_xlabel('month_number')
ax1.set_ylabel('total profit')
ax1.set_title("total profit of all months per month")
var.plot(kind='bar')
plt.show()
```

### 2) sales of all different products per product per month in one plot

```python

#In this example I used line plot, because we have different products and sales that change each month for each product,
#and line plot is most frequently used to analyses how the data has changed over time. here the purpose is to make comparison
#between sales of different products.

import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/home/depinfo/Documents/DATA SCIENCE/Visualization TP-20200403/company_sales_data.csv", sep=',') 
print(df)

var1 = df.groupby(['month_number']).facecream.sum()
var2 = df.groupby(['month_number']).facewash.sum()
var3 = df.groupby(['month_number']).toothpaste.sum()
var4 = df.groupby(['month_number']).bathingsoap.sum()
var5 = df.groupby(['month_number']).shampoo.sum()
var6 = df.groupby(['month_number']).moisturizer.sum()


fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.set_xlabel('Months')
ax1.set_ylabel('Sales')
ax1.set_title("sales of all different products per product per month")

var1.plot(kind='line',label="facecream")
var2.plot(kind='line', label="facewash")
var3.plot(kind='line', label="toothpaste")
var4.plot(kind='line', label="bathingsoap")
var5.plot(kind='line', label="shampoo")
var6.plot(kind='line', label="moisturizer")

ax1.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

plt.show()
```

### 3) compare face cream vs. toothpaste sales per month 

```python

#In this example I decided to use two different method and technics to plot : using Line plot and heat Map 

#Heat map : Heatmaps are good for showing variance across multiple variables,displaying whether any variables are similar
#to each other, and for detecting if any correlations exist in-between them, here we want to compare between 
#face cream and toothpaste sales per month 

#Line plot : We can also use line plot, which is good for comparison 

import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/home/depinfo/Documents/DATA SCIENCE/Visualization TP-20200403/company_sales_data.csv", sep=',') 

var1 = df.groupby(['month_number']).facecream.sum()
var2 = df.groupby(['month_number']).toothpaste.sum()

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.set_xlabel('month_number')
ax1.set_ylabel('Sales')
ax1.set_title("face cream vs. toothpaste sales per month ")

var1.plot(kind='line',label="facecream")
var2.plot(kind='line', label="toothpaste")

ax1.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

plt.show()
```

### 4) distribution of total profits

```python
#I used Histogram because Histograms help give an estimate as to where values are concentrated, 
#what the extremes are and whether there are any unusual values. and Here we need to have the distribution of
#of total profit which is a continuous interval, that's why histogramm is more appropiate in this case.



fig=plt.figure() #Plots in matplotlib reside within a figure object, use plt.figure to create new figure
#Create one or more subplots using add_subplot, because you can't create blank figure
ax = fig.add_subplot(1,1,1)
#Variable
ax.hist(df['total_profit'],bins = 10) 
# Here you can play with number of bins Labels and Tit
plt.title('Total profit distribution')
plt.xlabel('total profits')
plt.ylabel('#Product')
plt.show()
```

### 5) product sales of each product and a cumulative view of all products in one plot 

```python
#I used Box plot, is a convenient way of visually displaying the data distribution, we can see how the sales distributes themselves 
#from one product to another, and also the sales of each product, we have a global vision of all this in one plot 
#that's why box plot is most suitable in this case

fig=plt.figure()
ax = fig.add_subplot(1,1,1)
#Variable
plt.boxplot([df['facecream'], df['facewash'], df['toothpaste'], df['bathingsoap'], df['shampoo'], df['moisturizer']])
plt.title('product sales of each product and a cumulative view of all products')
plt.xticks( [1, 2, 3, 4, 5, 6],['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer'])
plt.ylabel('Sales')

plt.show()
```
