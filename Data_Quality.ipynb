```python
# r'^(\d{4})': The regular expression is meant to find any four digits at the beginning of a string, the purpose 
# here is to be able to convert these dates into numerical data

extr = df2['start_date'].str.extract(r'^(\d{4})', expand=False)
extr.head(8)
```

```python
# this output column still has object dtype, but we can easily get its numerical using pd.to_numeric
df2['start_date'] = pd.to_numeric(extr)
df2['start_date'].dtype                
```

```python
df2['start_date'] = pd.to_datetime(df2['start_date'])
df2['start_date'] = df2['start_date'].apply(lambda x: x.date())
df2['start_date'].dtype
```
